```python
#Noting the time of transaction
import time

#Reading the data form "Inventory.txt"
fd = open('Inventory.txt','r')
products = fd.read().split('\n')
fd.close()

#Taking User's input
ui_username = input("Enter Name : ")
ui_phone = input("Enter phone number : ")
ui_mail = input("Enter mail id : ")
ui_product_id = input("Enter the product id : ")
ui_product_qn = input("Enter the Qantity  : ")

update_product_lst = []
#Going through each product details
for product in products:
  prod_details = product.split(',')

  if (prod_details[0] == ui_product_id):
  #Checking if the asked product's quantity is available or not
    if(int(ui_product_qn) <= int(prod_details[3])):
      print("Product is available")
      print('*'*50)
      print("Product name               :", prod_details[1])
      print("Product Price              :", prod_details[2])
      print("Available Product Quantity :", prod_details[3])
      print("Product Quantity  want(s)  :", ui_product_qn)
      print('*'*50)
      print("Billing amount             :", int(ui_product_qn ) * int(prod_details[2]) )
      print('*'*50)
      print(f"Thanks for shopping {ui_username}, Visit Again.")
      print('*'*50)
      #Updatung Inventory list
      prod_details[3] = str(int(prod_details[3]) - int(ui_product_qn ))
# Generating sales in 'sales.txt'
      fd = open("sales.txt", 'a')
      sales_details =  ui_username + ',' +  ui_phone + ',' + ui_mail +  ',' + ui_product_id +  ',' + prod_details[1] + ',' + ui_product_qn + ','+ str( int(ui_product_qn ) * int(prod_details[2])) + ',' + time.ctime() + '\n'
      fd.write(sales_details)
      fd.close()



      print(f"Remaining quantity of the product {ui_product_id} is {prod_details[3]}")
      print('*'*50)
    else:
      if(int(prod_details[3]) > 0):

        #If we're not having enough quantity
        print("Sorry, we're not having enough Quantity.")
        print(f"We're having only {prod_details[3]} Available currently.")
        print("Would you like to Purchase it?")


        #If buyer wants to purchase the available quantity
        ch = input("Enter y/n : ")
        if (ch == 'y'):

          print("Product is available")
        print('*'*50)
        print("Product name               :", prod_details[1])
        print("Product Price              :", prod_details[2])
        print("Product Quantity  want(s)  :", ui_product_qn)
        print("Available Product Quantity :", prod_details[3])
        print('*'*50)
        print("Billing amount             :", int(prod_details[3]) * int(prod_details[2]) )
        print('*'*50)
        print(f"Thanks for shopping {ui_username}, Visit Again.")
        print('*'*50)
      #Updatung Inventory list
      prod_details[3] = '0'
      print(f"Remaining quantity of the product {ui_product_id} is {prod_details[3]}")
      print('*'*50)
      
      if(int(prod_details[3]) == 0):
        print("sorry, we're out of stock.")
        print('*'*50)
        print(f"Thanks for Visiting {ui_username}, We'll inform you via call/email, when the product {prod_details[1]} is available.")
        print('*'*50)

      # Generating sales in 'sales.txt'
      fd = open("sales.txt", 'a')
      sales_details =  ui_username + ',' +  ui_phone + ',' + ui_mail +  ',' + ui_product_id  +  ',' + prod_details[1] + ',' + prod_details[3] + ','+ str( int(ui_product_qn ) * int(prod_details[2])) + ',' + time.ctime() + '\n'
      fd.write(sales_details)
      fd.close()

  update_product_lst.append(prod_details)

#Updating my inventory list

lst = []

for i in update_product_lst:
  prod = i[0] + ',' + i[1] + ',' + i[2] + ',' + i[3] + '\n'
  lst.append(prod)

#Removing last \n from lst

lst[-1] = lst[-1][:-1]
#Updating Inventory file
fd = open( 'Inventory.txt','w')

for i in lst:
  fd.write(i)

fd.close()
```
